```python
# Ground truth
f = open("ground.in", "w+")
for x in range(100):
    f.write(' {:0d}\n'.format(0))

for x in range(100):
    f.write(' {:0d}\n'.format(1))

f.close()
# Convert the floating text in the file to integers when they come in
ground = np.loadtxt("ground.in").astype(np.int32)
#print(ground)
```

```python
# Make some toy data sets
# Set 1: 2 horizontal lines: 0 < y < 1
count = 0.1
f = open("lines-1.in", "w")
for x in range(100):
    f.write('{:2.2f}'.format(count))
    f.write(' {:0d}\n'.format(0))
    count += 0.1
    
count = 0.1
for x in range(100):
    f.write('{:2.2f}'.format(count))
    f.write(' {:0d}\n'.format(1))
    count += 0.1

f.close()
lines1 = np.loadtxt("lines-1.in")

# K-means
kmeans = KMeans(n_clusters=2).fit(lines1)
#print(kmeans)
#print(kmeans.cluster_centers_)
#print(kmeans.labels_)

# Plot the K-means attempt at the data
# Red will correlate to all 0's seen in kmeans.labels_, blue correlates to all 1's.
plt.scatter(lines1[:,0], lines1[:,1], color = [["red", "blue"][i] for i in kmeans.labels_])
plt.ylabel('Y')
plt.xlabel('X')
plt.show()

# Perform NMI
from sklearn.metrics.cluster import normalized_mutual_info_score
normalized_mutual_info_score(ground, kmeans.labels_).round(10)
```

```python
# Make some toy data sets
# Set 4: Interlocking Circles
# Make one circle

pi = math.pi

def PointsInCircum(r,n):
    return [(math.cos(2*pi/n*x)*r,math.sin(2*pi/n*x)*r) for x in range(0,n+1)]

f = open("circle-1.in", "w")

circle1 = PointsInCircum(2, 100)

for x in range(100):
    f.write('{:0f} '.format(circle1[x][0]))
    f.write(' {:0f}\n'.format(circle1[x][1]))
    
f.close()
circle1 = np.loadtxt("circle-1.in")
#print(circle1)

# Make one circle shifted to the right.

def PointsInCircum_shift_right(r,n):
    return [(math.cos(2*pi/n*x)*r + 2,math.sin(2*pi/n*x)*r) for x in range(0,n+1)]

f = open("circle-2.in", "w")

circle2 = PointsInCircum_shift_right(2, 100)

for x in range(100):
    f.write('{:0f} '.format(circle2[x][0]))
    f.write(' {:0f}\n'.format(circle2[x][1]))
    
f.close()
circle2 = np.loadtxt("circle-2.in")

# Bring the two circles together

f = open("interlocking_circles.in", "w")

for x in range(100):
    f.write('{:0f} '.format(circle1[x][0]))
    f.write(' {:0f}\n'.format(circle1[x][1]))
    
for x in range(100):
    f.write('{:0f} '.format(circle2[x][0]))
    f.write(' {:0f}\n'.format(circle2[x][1]))
    
f.close()
interlocking_circles = np.loadtxt("interlocking_circles.in")

# K-means
kmeans = KMeans(n_clusters=2).fit(interlocking_circles)
#print(kmeans.cluster_centers_)
#print(kmeans.labels_)

# Plot the K-means attempt at the data
plt.scatter(interlocking_circles[:,0], interlocking_circles[:,1], color = [["red", "blue"][i] for i in kmeans.labels_])
plt.ylabel('Y')
plt.xlabel('X')
plt.show()

# Perform NMI
from sklearn.metrics.cluster import normalized_mutual_info_score
normalized_mutual_info_score(ground, kmeans.labels_).round(10)
```

```python
# Make some toy data sets
# Set 5: Interlocking U's 
# top half

def PointsInCircum_shift_up(r,n):
    return [(math.cos(2*pi/n*x)*r,math.sin(2*pi/n*x)*r + 3) for x in range(0,n+1)]

f = open("u-1.in", "w")

u1 = PointsInCircum_shift_up(2, 200)

for x in range(100):
    f.write('{:0f} '.format(u1[x+100][0]))
    f.write(' {:0f}\n'.format(u1[x+100][1]))
    
f.close()
u1 = np.loadtxt("u-1.in")

# Interlocking U's: bottom half

def PointsInCircum(r,n):
    return [(math.cos(2*pi/n*x)*r,math.sin(2*pi/n*x)*r) for x in range(0,n+1)]

f = open("u-2.in", "w")

u2 = PointsInCircum(2, 200)

for x in range(100):
    f.write('{:0f} '.format(u2[x][0]))
    f.write(' {:0f}\n'.format(u2[x][1]))
    
f.close()
u2 = np.loadtxt("u-2.in")

# Interlocking U's: Combine the upper and lower half.

f = open("interlocking_us.in", "w")

for x in range(100):
    f.write('{:0f} '.format(u1[x][0]))
    f.write(' {:0f}\n'.format(u1[x][1]))
    
for x in range(100):
    f.write('{:0f} '.format(u2[x][0]))
    f.write(' {:0f}\n'.format(u2[x][1]))
    
f.close()
interlocking_us = np.loadtxt("interlocking_us.in")

# K-means
kmeans = KMeans(n_clusters=2).fit(interlocking_us)
#print(kmeans.cluster_centers_)
#print(kmeans.labels_)

# Plot the K-means attempt at the data
plt.scatter(interlocking_us[:,0], interlocking_us[:,1], color = [["red", "blue"][i] for i in kmeans.labels_])
plt.ylabel('Y')
plt.xlabel('X')
plt.show()

# Perform NMI
from sklearn.metrics.cluster import normalized_mutual_info_score
normalized_mutual_info_score(ground, kmeans.labels_).round(10)
```

```python
# Make some toy data sets
# Set 6: 2 lines making an X
count = -50
f = open("x-1.in", "w")
for x in range(100):
    f.write('{:2.2f}'.format(count))
    f.write(' {:2.2f}\n'.format(count))
    count += 1
    
count = -50
y = 0
for x in range(100):
    y = (-1)*(count)
    f.write('{:2.2f}'.format(count))
    f.write(' {:2.2f}\n'.format(y))
    count += 1

f.close()
x1 = np.loadtxt("x-1.in")

# K-means
kmeans = KMeans(n_clusters=2).fit(x1)
#print(kmeans.cluster_centers_)
#print(kmeans.labels_)

# Plot the K-means attempt at the data
plt.scatter(x1[:,0], x1[:,1], color = [["red", "blue"][i] for i in kmeans.labels_])
plt.ylabel('Y')
plt.xlabel('X')
plt.show()

# Perform NMI
from sklearn.metrics.cluster import normalized_mutual_info_score
normalized_mutual_info_score(ground, kmeans.labels_).round(10)
```
